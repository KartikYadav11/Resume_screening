```python
def cleanResume(resumeText):
    resumeText = re.sub('httpS+s*', ' ', resumeText)
    resumeText = re.sub('RT|cc', ' ', resumeText) 
    resumeText = re.sub('#S+', '', resumeText)  
    resumeText = re.sub('@S+', '  ', resumeText)  
    resumeText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[]^_`{|}~"""), ' ', resumeText)  
    resumeText = re.sub(r'[^x00-x7f]',r' ', resumeText) 
    resumeText = re.sub('s+', ' ', resumeText)  
    return resumeText
```

```python
data['cleaned_resume'] =data.Resume.apply(lambda x: cleanResume(x))
```

```python
data.head(3)
```

```python
var_mod=['Category']
le=LabelEncoder()
for i in var_mod:
    data[i]=le.fit_transform(data[i])
```

```python
data['Category']
```

```python
requiredText = data['cleaned_resume'].values
```

```python
requiredTarget = data['Category'].values
```

```python
word_vectorizer=TfidfVectorizer(
     sublinear_tf=True,stop_words='english',max_features=1000)
```

```python
word_vectorizer.fit(requiredText)
```

```python
WordFeatures = word_vectorizer.transform(requiredText)
```

```python
X_train,X_test,y_train,y_test = train_test_split(WordFeatures,requiredTarget,random_state=101, test_size=0.3)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
clf = KNeighborsClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
prediction = clf.predict(X_test)
```

```python
print('Accuracy of KNeighbors Classifier on training set: {:.2f}'.format(clf.score(X_train, y_train)))
print('Accuracy of KNeighbors Classifier on test set: {:.2f}'.format(clf.score(X_test, y_test)))
```
